```python
from mido import MidiFile
```

```python

mid = MidiFile('division-violin-two-27.mid', clip=False)
```

```python
print(mid)
```

```python
for track in mid.tracks:
    print(track)
```

```python
print(len(mid.tracks))
```

```python
for track in mid.tracks:
    for msg in track:
        if (msg.type == 'note_on'):
            print(msg.note)
            print(msg)
```
