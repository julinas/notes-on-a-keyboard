```python
from bs4 import BeautifulSoup
from os import path
import requests
```

```python
def download_url(url, save_path, chunk_size=128):
    r = requests.get(url, stream=True)
    with open(save_path, 'wb') as fd:
        for chunk in r.iter_content(chunk_size=chunk_size):
            fd.write(chunk)
```

```python
def get_all_links_on_page(soup):
    links = soup.find_all("a")
    for l in links:
        if "mid file" in l.text:
            url = l['href']
            target_path = "data/" + path.basename(url)
            download_url(url, target_path)
```

```python
def get_next_page(soup):
    links = soup.find_all("a")
    for l in links:
        if "Next 10" in l.text:
            url = "https://www.mutopiaproject.org/cgibin/" + l['href']
            print(url)
            page = requests.get(url, timeout=5)
            soup = BeautifulSoup(page.content, "html.parser")
            return soup
    return None
```

```python
# url = "https://www.mutopiaproject.org/cgibin/make-table.cgi?Style=Baroque"
styles = ["Classical", "Folk", "Gospel", "Hymn", "Jazz", "March", "Modern", "Popular%20/%20Dance", "Renaissance", "Romantic", "Song", "Technique"]
urls = ["https://www.mutopiaproject.org/cgibin/make-table.cgi?Style={}".format(x) for x in styles]
for url in urls:
    page = requests.get(url, timeout=5)
    soup = BeautifulSoup(page.content, "html.parser")
    while soup is not None:
        get_all_links_on_page(soup)
        soup = get_next_page(soup)   
```
